Question 9 ----> CLASSIFY THE MOVIES BASED ON RATINGS ( EXCELLENT, GOOD OR AVERAGE)

```python
# we going to use if and elif function.ab

def Ratingss(ratings):
    if ratings>=7.0:
        return "Excellent"
    elif ratings>=5.0:
        return "Good"
    else:
        return "Average"
```

```python
imdb_data["Ratings_type"] = imdb_data["Rating"].apply(Ratingss)
```

```python
imdb_data.head(5)
```

AS WE CAN SEE THERE IS ANOTHER COLUMN ADDED "Ratings_Type" THAT DEFINES IS THAT MOVIE
IS EXCELLENT , GOOD OR AVERAGE.
